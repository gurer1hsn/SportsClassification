```
!pip install tensorflow gitpython
from google.colab import drive
drive.mount('/content/drive')

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from git import Repo
from getpass import getpass

data_dir = '/content/drive/MyDrive/SportsDataset'

# Dizinin var olup olmadığını kontrol et
if not os.path.exists(data_dir):
    print(f"Dizin mevcut değil: {data_dir}")
else:
    print(f"Dizin mevcut: {data_dir}")

# ImageDataGenerator ve flow_from_directory kullanarak veri setini yükle
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='training')

validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation')

steps_per_epoch = np.ceil(train_generator.samples / train_generator.batch_size)
validation_steps = np.ceil(validation_generator.samples / validation_generator.batch_size)

# CNN modeli oluşturma ve eğitme
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    epochs=15)

# Modeli kaydetme
model.save('/content/drive/MyDrive/SportsDataset/sports_classification_model.h5')

# Create the directory if it doesn't exist
if not os.path.exists('/content/SportsClassification'):
    os.makedirs('/content/SportsClassification')

# .gitignore dosyasını ekleyin
with open('/content/SportsClassification/.gitignore', 'w') as f:
    f.write('*.h5\n')

# Doğru dizinde olduğunuzdan emin olun
os.chdir('/content/SportsClassification')

# Kullanıcıdan GitHub token'ını gizli bir şekilde alma
github_token = getpass('Enter your GitHub token: ')

# Kişisel erişim belirteci ile git config ayarları
!git config --global user.email "hasanhuseyingurer06@gmail.com"
!git config --global user.name "gurer1hsn"

# Git deposunu başlat ve değişiklikleri yükle
!git init
!git remote add origin https://{github_token}@github.com/gurer1hsn/SportsClassification.git
!git add .
!git commit -m "Modeli ve eğitim çıktılarını ekledim"
!git push -u origin main
```
